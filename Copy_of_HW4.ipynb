```python
import pandas as pd
from scipy.stats import chi2_contingency


# Define the target variable
target_variable = 'LEAVE'

# Define the categorical columns you want to test for correlation
categorical_columns_to_test =['COLLEGE', 'INCOME', 'OVERAGE', 'LEFTOVER', 'HOUSE', 'HANDSET_PRICE', 'OVER_15MINS_CALLS_PER_MONTH', 'AVERAGE_CALL_DURATION', 'REPORTED_SATISFACTION', 'REPORTED_USAGE_LEVEL', 'CONSIDERING_CHANGE_OF_PLAN', 'LEAVE']

# Create a contingency table for each categorical variable
correlation_results = {}

for column in categorical_columns_to_test:
    contingency_table = pd.crosstab(Data[target_variable], Data[column])
    chi2, p, _, _ = chi2_contingency(contingency_table)
    correlation_results[column] = {'Chi-Squared': chi2, 'P-Value': p}

# Print the correlation results
print("Correlation Analysis:")
for column, results in correlation_results.items():
    print(f"{column}:")
    print(f"Chi-Squared: {results['Chi-Squared']:.2f}")
    print(f"P-Value: {results['P-Value']:.4f}")
    if results['P-Value'] < 0.05:
        print("Statistically significant correlation.")
    else:
        print("No statistically significant correlation.")
    print()
```

K-means Method

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Assuming 'Data' is your DataFrame with relevant columns

# Selecting numerical columns for clustering
numerical_columns_for_clustering = ['INCOME', 'OVERAGE', 'LEFTOVER', 'HOUSE', 'HANDSET_PRICE', 'OVER_15MINS_CALLS_PER_MONTH', 'AVERAGE_CALL_DURATION']

# Extracting the relevant data
cluster_data = Data[numerical_columns_for_clustering]

# Standardizing the data (important for k-means)
cluster_data_standardized = (cluster_data - cluster_data.mean()) / cluster_data.std()

# Determine the optimal number of clusters (k) using the Elbow Method
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10, max_iter=300).fit(cluster_data_standardized)
    inertia.append(kmeans.inertia_)

# Plotting the Elbow curve
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.show()

# Based on the Elbow Method, choose the optimal value of k
optimal_k = 3  # Update with the value you observe from the plot

# Fit the k-means model with the optimal k
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10, max_iter=300)
cluster_data['Cluster'] = kmeans.fit_predict(cluster_data_standardized)

# Calculate the centroids using Euclidean distance in the original scale
centroids = []
for cluster_label in range(optimal_k):
    cluster_points = cluster_data[cluster_data['Cluster'] == cluster_label][numerical_columns_for_clustering]
    centroid = cluster_points.mean()
    centroids.append(centroid)

# Visualize the clustering using a scatter plot
plt.figure(figsize=(12, 8))
sns.scatterplot(x='INCOME', y='OVERAGE', hue='Cluster', data=cluster_data, palette='viridis', s=100)
centroids = pd.DataFrame(centroids, columns=numerical_columns_for_clustering)
plt.scatter(centroids['INCOME'], centroids['OVERAGE'], c='red', marker='X', s=200, label='Centroids')
plt.title('K-Means Clustering of Customers with Centroids (Euclidean Distance)')
plt.xlabel('Income')
plt.ylabel('Overage')
plt.legend()
plt.show()

# Analyze the centroids
print("Cluster Centers (Centroids):")
print(centroids)

# Analyze the statistics of variables within the clusters compared to the whole dataset
cluster_summary = cluster_data.groupby('Cluster').describe().transpose()

# Visualize the cluster profiles
plt.figure(figsize=(12, 8))
sns.heatmap(centroids.T, annot=True, cmap="YlGnBu", fmt=".2f", linewidths=.5)
plt.title('Cluster Centers (Centroids)')
plt.show()

# Display statistics of variables within clusters compared to the whole dataset
print("Cluster Summary:")
print(cluster_summary)
```

```python
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt

# Create and train a Logistic Regression model
logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)

# Extract coefficients
coefficients = logistic_model.coef_[0]
coefficients_dict = dict(zip(X.columns, coefficients))

# Print or plot coefficients
print("Coefficients:")
for feature, coefficient in coefficients_dict.items():
    print(f"{feature}: {coefficient:.4f}")

# Plotting coefficients
sorted_coefficients = sorted(coefficients_dict.items(), key=lambda x: abs(x[1]), reverse=True)
features, coefficients = zip(*sorted_coefficients)

plt.figure(figsize=(10, 6))
plt.barh(range(len(features)), coefficients, align="center")
plt.yticks(range(len(features)), features)
plt.xlabel("Coefficient Value")
plt.title("Logistic Regression - Coefficients")
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt

# Create and train a KNN model
knn_model = KNeighborsClassifier()
knn_model.fit(X_train, y_train)

# KNN doesn't have inherent feature importances, but you can use feature importance based on mutual information
from sklearn.feature_selection import mutual_info_classif

# Calculate mutual information between each feature and the target variable
mutual_info = mutual_info_classif(X_train, y_train)

# Create a dictionary with feature names and their mutual information scores
mutual_info_dict = dict(zip(X.columns, mutual_info))

# Print or plot mutual information scores
print("Mutual Information Scores:")
for feature, info_score in mutual_info_dict.items():
    print(f"{feature}: {info_score:.4f}")

# Plotting mutual information scores
sorted_info_scores = sorted(mutual_info_dict.items(), key=lambda x: x[1], reverse=True)
features, info_scores = zip(*sorted_info_scores)

plt.figure(figsize=(10, 6))
plt.barh(range(len(features)), info_scores, align="center")
plt.yticks(range(len(features)), features)
plt.xlabel("Mutual Information Score")
plt.title("K-Nearest Neighbors - Mutual Information Scores")
plt.show()
```
